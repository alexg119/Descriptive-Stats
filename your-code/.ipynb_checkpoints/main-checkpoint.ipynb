## Challenge 1
#### 1.- Define a function that simulates rolling a dice 10 times. Save the information in a dataframe.
**Hint**: you can use the *choices* function from module *random* to help you with the simulation.

```python
# your code here
def dice_roll(num_rolls):
    results = []  
    for i in range(num_rolls):  
        roll = random.randint(1, 6)  
        results.append(roll)  
    return results

rolls = dice_roll(10)

df = pd.DataFrame({'result': rolls}) 

print(df)
    
```

#### 2.- Plot the results sorted by value.

```python
# your code here
df_sorted = df.sort_values(by="result", inplace=False)

# Reset the index 
df_sorted = df_sorted.reset_index(drop=True)

#plot using the reset index to have it sorted by value
graph = df_sorted.reset_index().plot.scatter(x='index', y='result', c='orange', title='Dice Roll Results')

# Display the plot
plt.show()
```

#### 3.- Calculate the frequency distribution and plot it. What is the relation between this plot and the plot above? Describe it with words.

```python
# your code here
#Let's make a histogram using a bin for each result

ax = df_sorted['result'].plot.hist(bins=6, alpha=0.75, title='Frequency Distribution of Dice Rolls')

# Set the labels
ax.set_xlabel('Dice Result')
ax.set_ylabel('Frequency')

plt.show()
```

```python
"""
The histogram doesn't care about the order of the dice rolls, it is simply showing how often each result was achieved. The scatterplot on 
the other hand does show an order which can be used (not in this case with dice rolls) to show potential patterns between the two axis.
Both plots convey the same data put present it differently. If we saw a scatterplot with a condensed area of results the histogram would reflect
this as well with a weighted frequency distribution
"""
```

## Challenge 2
Now, using the dice results obtained in *challenge 1*, your are going to define some functions that will help you calculate the mean of your data in two different ways, the median and the four quartiles. 

#### 1.- Define a function that computes the mean by summing all the observations and dividing by the total number of observations. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
# your code here
def mean(rolls):
    mean = sum(rolls) / len(rolls)
    return mean

print(mean(rolls))
```

#### 2.- First, calculate the frequency distribution. Then, calculate the mean using the values of the frequency distribution you've just computed. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
# your code here
freq_dist = {}

def calc_freq_dist(x):
    freq_dist_calc = {}
    for num in x:
        if num in freq_dist_calc:
            freq_dist_calc[num] += 1
        else:
            freq_dist_calc[num] = 1
    return freq_dist_calc

freq_dist = calc_freq_dist(rolls)

print(freq_dist)

#In hindsight, working with a dictionary comprehension is going to get quite annoying If I were to do it again
#I would just create a list from the created DataFrame


values = df_sorted['result'].tolist()

print(values)

total_sum = 0

def calc_sum(x):
    total_sum = 0
    for key, value in x.items():
        product = key * value
        total_sum += product
    return total_sum

print(calc_sum(calc_freq_dist(rolls))/len(rolls))
```

#### 3.- Define a function to calculate the median. You are not allowed to use any methods or functions that directly calculate the median value. 
**Hint**: you might need to define two computation cases depending on the number of observations used to calculate the median.

```python
#Actually now I'm not sure, its quite easy to just create the list within in each function - I'm not sure what the best practice is here

def median(x):
    #Adding check depending on input list vs dict
    if isinstance(x, dict):
        flat_list = []
        for key, value in x.items():
            flat_list.extend([key] * value)
    else:
        flat_list = list(x)

    #sort the list - dict will already be sorted
    flat_list.sort()
    
    length = len(flat_list)
    mid = length // 2
    if length % 2 == 0:
        return (flat_list[mid - 1] + flat_list[mid]) / 2
    else:
        return flat_list[mid]

print(median(freq_dist))

print(median(values))
```

#### 4.- Define a function to calculate the four quartiles. You can use the function you defined above to compute the median but you are not allowed to use any methods or functions that directly calculate the quartiles. 

```python
# your code here
def quartiles(x):
    if isinstance(x, dict):
        flat_list = []
        for key, value in x.items():
            flat_list.extend([key] * value)
        flat_list.sort()
    else:
        flat_list = sorted(x)
    
    length = len(flat_list)
    #lets split the list in half

    if length % 2 == 0:
        lower_half = flat_list[:length // 2]
        upper_half = flat_list[length // 2:]
    else:
        lower_half = flat_list[:length // 2]
        upper_half = flat_list[length // 2 + 1:]

    Q2 = median(flat_list)
    Q1 = median(lower_half)
    Q3 = median(upper_half)

    Quartiles = [Q1, Q2, Q3]
    
    return Quartiles

print(quartiles(freq_dist))
print(quartiles(values))

    
```
